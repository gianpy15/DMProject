```python
def expand_datetime_utc(df):
    res_df = df.copy()
    res_df = res_df.reset_index()

    res_df = pd.DataFrame({
        col: np.repeat(res_df[col].values, res_df.DATETIME_UTC.str.len())
        for col in res_df.columns.drop('DATETIME_UTC')}
    ).assign(**{'DATETIME_UTC': np.concatenate(res_df.DATETIME_UTC.values)})[res_df.columns]

    return res_df
```

```python
dataset = data.base_dataset('full','test')
lorofile = pd.read_csv('resources/dataset/originals/speeds_evaluation__only_datetimes__2019.csv')
lorofile['DATETIME_UTC'] = pd.to_datetime(lorofile['DATETIME_UTC'])

check_missing0 = lorofile[lorofile.PREDICTION_STEP == 0].merge(dataset[['KEY','KM','event_index','DATETIME_UTC_y_0']], how='left', \
left_on=['KEY','KM','DATETIME_UTC'], right_on=['KEY','KM','DATETIME_UTC_y_0'])
check_missing1 = lorofile[lorofile.PREDICTION_STEP == 1].merge(dataset[['KEY','KM','event_index','DATETIME_UTC_y_1']], how='left', \
left_on=['KEY','KM','DATETIME_UTC'], right_on=['KEY','KM','DATETIME_UTC_y_1'])
check_missing2 = lorofile[lorofile.PREDICTION_STEP == 2].merge(dataset[['KEY','KM','event_index','DATETIME_UTC_y_2']], how='left', \
left_on=['KEY','KM','DATETIME_UTC'], right_on=['KEY','KM','DATETIME_UTC_y_2'])
check_missing3 = lorofile[lorofile.PREDICTION_STEP == 3].merge(dataset[['KEY','KM','event_index','DATETIME_UTC_y_3']], how='left', \
left_on=['KEY','KM','DATETIME_UTC'], right_on=['KEY','KM','DATETIME_UTC_y_3'])

print('missing step 0:', sum(check_missing0.event_index.isnull()*1))
print('missing step 1:', sum(check_missing1.event_index.isnull()*1))
print('missing step 2:', sum(check_missing2.event_index.isnull()*1))
print('missing step 3:', sum(check_missing3.event_index.isnull()*1))
```
